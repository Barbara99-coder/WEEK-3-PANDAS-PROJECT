IMPORTING DATA

TELCOM DATASET 2

```python
Telcom2 = pd.read_csv("/content/Telcom_dataset2.csv")
Telcom2
```

TELCOM DATASET 3

```python
Telcom3 = pd.read_csv("/content/Telcom_dataset3.csv")
Telcom3
```

DATA CLEANING

```python
#renaming columns in telcom table 1

Telcom1.rename(columns={'PRODUTC':'PRODUCT','DATETIME':'DATE_TIME'},inplace=True)
Telcom1
```

MAKING APPROPRIATE COLUMNS FOR TELCOM DATA 2

```python

Telcom2.rename(columns={'DW_A_NUMBER':'DW_A_NUMBER_INT','DW_B_NUMBER':'DW_B_NUMBER_INT'},inplace=True)
Telcom2
```

MAKING APPROPRIATE COLUMNS FOR DATASET 3 

```python
Telcom3.rename(columns={'CELLID':'CELL_ID','SIET_ID':'SITE_ID'},inplace=True)
Telcom3
```

DROPPING NULL VALUES AND DUPLICATED VALUES FROM EACH TELCOM TABLE

```python
# TELCOM1
Telcom1 = Telcom1.dropna(axis=0,how='any')
Telcom1

Telcom1.drop_duplicates(keep=False,inplace=True)
Telcom1
```

```python
# TELCOM 2
Telcom2 = Telcom2.dropna(axis=0,how='any')

Telcom2.drop_duplicates(keep=False,inplace=True)
Telcom2
```

```python
# TELCOM 3
Telcom3 = Telcom3.dropna(axis=0,how='any')

Telcom3.drop_duplicates(keep=False,inplace=True)
Telcom3
```

MERGING THE THREE TELCOM DATAFRAMES

```python
Merged_telcom_data = pd.concat((Telcom1,Telcom2,Telcom3),axis=0,ignore_index=True)
Merged_telcom_data
```
